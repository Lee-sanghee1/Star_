#### 1. 언어

 - 인공지능의 분야 중 언어의 분야에서 현재 활발히 서비스를 진행하고 있는 ‘papago’는 인공신경망을 기반으로 만들어진 네이버의 대표 번역기이다.

 기존의 구글이나 네이버에서 이용하던 번역방식은 SMT(Statistical Machin Translation)로, 대규모 데이터를 기반으로 자동으로 학습된 통계적 모델을 기반으로 번역한다. 기본적으로는 단위마다 번역해서 조합하는 방식이지만, 구(Phrase) 형식 기반 번역으로 발전했다. 구 단위가 단어 단위보다 모호성이 낮기 때문이다. 구를 기반으로 하면 더 나은 번역 품질을 얻을 수 있다.

 하지만 여기서 한 단계 발전한 NMT(Neural Machine Translation) 번역방식은 인공 신경망 기계번역을 뜻한다. 문장 전체를 넣어서 번역하는 방식으로, 핵심은 단어 표현, 즉 워드 임베딩(word embedding)이다. 단어 표현이란 하나의 단어를 인공 신경망을 이용하여 벡터 공간상에 나타낼 수 있는 변환된 값을 의미한다. 번역되는 방식을 예로 들면 한 입체의 공간에 하나의 단어를 띄우고 관계가 있는 단어들을 유사한 공간에 둔다. 또한 다른 차원에서 다른 단어들과 관계를 맺을 수 있어 공간 안에서 서로 매핑이 되며, 입력된 문장에서 번역하려는 문장으로의 최적의 가중치를 찾아 결과값을 구하는 방식이다.
 
 현재 파파고는 NMT방식에 더해 전체 번역 프레임워크를 딥러닝 방식으로 구축해 정확도를 2배 이상 높였다. 이에따라 문장의 자연스러움이나 은어, 유행어를 자연스럽게 번역하는데에는 구글번역기보다 나으나, 구글 대비 클라우드 서버 용량 부족으로 긴 문장의 번역에는 아직까지 흠이 있다고 한다.

#### 2. 음성

 - 빅스비는 삼성에서 자체 개발한 인공지능 음성인식 플랫폼이다. 삼성의 대표 갤럭시 시리즈에서 이용할 수 있다.

 이런 인공지능 음성인식의 가장 핵심기술은 NLP(Natural Language Processing), ‘자연어 처리 기술’과 ‘의미 검색’이다. 자연어 처리 기술(NLP)은 사람이 말하는 언어, 즉 자연어를 기계적으로 분석해 컴퓨터가 이해할 수 있는 형태로 만드는 기술을 의미한다. NLP 기술은 사람의 음성 언어를 컴퓨터가 이해하도록 만드는 것으로 STT(Speech to text)와는 구분된다. STT는 사람의 음성 언어를 문자 언어 형태로 컴퓨터가 변환하는 것만을 말하지만 NLP는 기계가 이해하기에는 직관적이고 모호한 표현이 많은 음성 언어를 컴퓨터가 이해하는데 그 목적이 있다. 또한 의미 검색 기술은 음성 언어를 기반으로 컴퓨터가 의미를 분석해 최적화된 결과를 검색하는 기술이다.

 하지만 AI 접목 이후 머신러닝 알고리즘 중 하나인 딥러닝의 개발로 NLP의 기술력 역시 한 단계 도약했다. 인공신경망 기반 딥러닝 알고리즘이 탑재된 AI는 데이터를 입력해 학습을 시키면 시킬수록 성능이 향상되는 현 모습을 비로소 갖추게 됐다. 특히 자연어 처리에 딥러닝 알고리즘이 적용되면서 기존에 단어 단위로 의미 분석을 했던 것에서 벗어나 구문 전체를 분석해 의미를 추론할 수 있는 수준으로 향상됐다.
 
 빅스비는 이런 딥러닝의 기술 적용으로 사용자와의 대화 내용과 일의 수행 처리 과정 또한 저장하며, 사용자가 긍정적 피드백을 한다면 처리 과정을 제대로 완수한 것으로 판단하고, 부정적 피드백을 한다면 어느 부분에서 문제가 생겼는지를 검토하고 오류를 수정한다. 

#### 3. 이미지

 - 구글포토는 촬영한 사진을 손쉽게 검색하고 친구, 가족과 공유할 수 있는 일종의 ‘사진 허브’이다. 그중 가장 핵심기능은 검색이며 인공지능 알고리즘을 통해 사물을 구분하는 단계에서 벗어나 사진의 배경, 위치, 자연현상, 이벤트 등을 스스로 인식하고 있다.

 구글 포토에 적용한 인공지능이 사용자가 원하는 사진을 찾기 위해서는 먼저 일단 구글에 업로드된 사진을 학습하는 단계에서 라벨(label) 작성 작업이 선행되야만 한다. 머신러닝 과정에서 사진을 인식해 사진속 사물이나 배경, 위치 등 다양한 정보를 추출하는 과정을 거쳐 검색이 가능하도록 일종의 검색 키워드를 생성하는 과정이 필요한데 이를 라벨이라 부른다. 보통 이미지당 10개 정도의 라벨이 생성되는데 이미지에 따라 생성되는 라벨의 종류가 다르다.
 
 물론 아무리 똑똑한 알고리즘을 추가하더라도 빈틈은 존재하기 마련이다. 구글이 이미지 머신러닝을 위해 사용하는 공개 이미지 데이터(open image dataset)는 900만개가 넘는데 그 중에 10만개를 골라 사람이 직접 검토하고 그 결과를 수정하는 방식으로 정확도를 높이고 있다.

#### 4. 자율주행

 - 전기차 생산으로 미래형 자동차 브랜드라는 이미지를 가지고있는 테슬라는 FSD(Full-Self Driving)라는 옵션을 만들어냈다. 이름 그대로 자율주행 자동차를 생산해낸 것인데, 이는 컴퓨터비전과 CNN(Convolutional Neural Network) 기술의 합작으로 가능해질 수 있다.

 CNN은 모델이 직접 이미지, 비디오, 텍스트 또는 사운드를 분류하는 머신 러닝의 한 유형인 딥러닝에 가장 많이 사용되는 알고리즘이다. 이미지에서 객체, 얼굴, 장면을 인식하기 위해 패턴을 찾는데 특히 유용하다. 이는 이미지의 형태 그대로의 pixel단위의 바이트배열을 그대로 학습시켜서 처리하여 학습에 따른 Layer구조를 설계해 학습처리만 하게 되면 이미지를 인식처리할 수가 있기 때문이다. 또한 최종출력단을 여러 개로 구별하도록 설계가 가능하므로 차량만을 인식하는 것이 아닌 다른 사물을 같이 인식 처리할 수 있다. 이것은 자율주행의 경우에 매우 장점이 되는 특성으로서 실질적으로 도로에서 다른 차량 뿐아니라 보행자, 자전거 등 여타 다른 사물을 인지하는 것이 중요한 문제이기 때문이다.
 
 이러한 신경망을 구축한 이후 컴퓨터 비전 즉, 컴퓨터가 사람 눈처럼 사물을 보고 이해하는 기술이 구축가능해 진 것이다. 카메라를 통해 자동차 전방, 후방 등 사방을 모두 인식하여 CNN 신경망을 통해 자율주행자동차가 속도를 늦춰야할지, 올려야할지, 차선을 변경해야할지 등을 판단하여준다.

 이렇게 조사한 여러 가지 기술 중 나는 CNN이라는 신경망에 관심이 생겼다. 왜냐하면 CNN은 텍스트 뿐만이 아니라 이미지, 비디오, 사운드 등 여러 자료들을 분류할 수 있다는 것에서 큰 매력을 느꼈기 때문이다. 데이터 분석가로서 웹상에서의 여러 데이터들을 추출하고 전처리하며 그 안에서의 인사이트를 도출하는 것이 큰 의미가 있는데, 딥러닝을 활용한 CNN은 내가 원하는 분석 기법을 활용하는데 있어 매우 활용도 높은 기술이다. 앞으로의 교육과정에 있어서 딥러닝의 원리를 파악하고 이를 데이터 분석하는데 있어 어떻게 활용할지 방안을 고안해내고싶다.
